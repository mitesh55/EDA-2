```python
# isilation forest :
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
data = pd.read_csv(r'IRIS.csv')
data.head()
```

```python
# onehotencodding on species data :
data_encoded = pd.get_dummies(data)
data_encoded.head()
```

```python
clf = IsolationForest(random_state=10, contamination=0.01)
clf.fit(data_encoded)
outlier_pred = clf.predict(data_encoded)
outlier_pred
```

```python
data_encoded["anomaly"] = pd.Series(outlier_pred)
# outlier_score = model.decision_function(data_encoded)
# outlier_score
```

```python
data_encoded.head()
```

```python
outlier_score = clf.decision_function(data_encoded)
outlier_score
```

```python
data_encoded["score"] = pd.Series(outlier_score)
data_encoded.head(10)
```
